### 匯入需要的函式庫

```python
import pandas as pd
import collections
```

### 讀取資料

```python
data = pd.read_excel('./1106_看門狗_銀行業分類整理-1.xlsx', index = False)
data.head()
```

### 選取需要的資料

```python
data = data[['公司', '事件日', '小事件類別', '事件內容']]
data.head()
```

### 依序按照　公司 / 事件日 / 小事件類別排序

```python
data.sort_values(by = ['公司', '事件日', '小事件類別'], inplace = True)
data.reset_index(inplace = True)
data.drop('index', axis = 1, inplace = True)
data.head()
```

### 將資料重新整理並合併

```python
group = data.groupby(by = ['公司', '事件日', '小事件類別'])
company = []
date = []
category = []
detail = []
for i in group.groups:
    company.append(i[0])
    date.append(i[1].date())
    category.append(i[2])
    text = list(map(str, group.get_group(i)['事件內容'].tolist()))
    detail.append('\n'.join(text))
```

### 將資料製作成表格

```python
combine = pd.DataFrame({'公司' : company,
                        '事件日' : date, 
                        '小事件類別' : category,
                        '事件內容' : detail})[['公司', '事件日', '小事件類別', '事件內容']]
combine.sort_values(by = ['公司', '事件日', '小事件類別', '事件內容'], inplace=True)
combine
```

### 輸出表格

```python
combine.to_excel('./1106_看門狗_銀行業分類整理-1test.xlsx', index = False)
```
